```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np

```

```python
city_data = "city_data.csv"
city_data_df = pd.read_csv(city_data)
city_data_df.head()
```

```python
ride_data = "ride_data.csv"
ride_data_df = pd.read_csv(ride_data)
ride_data_df.head()
```

```python
combined_city_data = pd.merge(ride_data_df, city_data_df, how='outer', on='city')
combined_city_data
```

```python
grouped_fare_avg = combined_city_data.groupby('city').mean()['fare']
grouped_rides = combined_city_data.groupby('city').count()['ride_id']
grouped_drivers = combined_city_data.groupby('city').count()['driver_count']

```

```python
urban_city = combined_city_data[combined_city_data['type']=='Urban']
suburban_city = combined_city_data[combined_city_data['type']=='Suburban']
rural_city = combined_city_data[combined_city_data['type']=='Rural']
# urban_percent = urban_city(len) / combined_city_data(len) * 100
rural_fares = rural_city['fare'].sum()
urban_fares = urban_city['fare'].sum()
suburban_fares = suburban_city['fare'].sum()
suburban_rides = len(suburban_city)
urban_rides = len(urban_city)
rural_rides = len(rural_city)
rural_drivers = rural_city['driver_count'].sum()
urban_drivers = urban_city['driver_count'].sum()
suburban_drivers = suburban_city['driver_count'].sum()
suburban_drivers





```

```python
plt.scatter(grouped_fare_avg, grouped_rides)
plt.title('Avg Fare')
plt.xlabel('Avg fare per trip')
plt.ylabel('Total Drivers')

plt.show()
```

```python


# doubling the width of markers

N = 50
area = (30 * np.random.rand(N))**2
colors = cm.rainbow(np.random.rand(N))
x = [grouped_rides]
y = [grouped_fare_avg]

plt.scatter(x,y,s=area, color=colors, alpha=0.5)

plt.xlabel('Total Rides')
plt.ylabel('Fare Avg')
plt.title('Trips')

plt.show
```

# Percent of fares by city type

```python
# Labels for the sections of our pie chart
labels = ['Urban Fares', 'Suburban Fares', 'Rural Fares']

sizes = [urban_fares, suburban_fares, rural_fares]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```
